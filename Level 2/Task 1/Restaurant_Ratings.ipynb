```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

# Load the dataset
data = pd.read_csv("/content/Dataset.csv")

# Convert data to DataFrame
df = pd.DataFrame(data)

# Display basic dataset information
print("ðŸ“Œ Dataset Information:")
df.info()

# Display the first 5 rows
print("\nðŸ“Œ First 5 Rows of the Dataset:")
print(df.head())

# Display the last 5 rows
print("\nðŸ“Œ Last 5 Rows of the Dataset:")
print(df.tail())

# Display summary statistics for numerical columns
print("\nðŸ“Œ Summary Statistics:")
print(df.describe())

# Check for missing values
print("\nðŸ“Œ Missing Values per Column:")
print(df.isnull().sum())

# Display column names
print("\nðŸ“Œ Column Names:")
print(df.columns.tolist())

# Display unique values for categorical columns (if applicable)
categorical_cols = df.select_dtypes(include=['object']).columns
print("\nðŸ“Œ Unique Values in Categorical Columns:")
for col in categorical_cols:
    print(f"{col}: {df[col].nunique()} unique values")

# Show shape of dataset (rows, columns)
print("\nðŸ“Œ Dataset Shape:")
print(df.shape)  # (rows, columns)

# Show data types
print("\nðŸ“Œ Data Types of Each Column:")
print(df.dtypes)
```

```python
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.histplot(df['Aggregate rating'], bins=range(1, 6), kde=False, color='skyblue', edgecolor='black')
plt.title('Distribution of Aggregate Ratings Among Restaurants')
```

```python
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
average_votes = df['Votes'].mean()
```

```python
print(f"Average Number of Votes Received by Restaurants: {average_votes:.2f}")
```
